# 2. Numerical Data

5. central tendancy
6. dispersion -> variance
7. shape

```python
print("mean=",df['Age'].mean())
print("var=",df['Age'].var())
print("median=",df['Age'].median())
print("skewness=",df['Age'].skew())
print("min=",df['Age'].min())
print("max=",df['Age'].max())
print("75 percentile=",df['Age'].quantile(0.75))
print("\n\n\n\n")
print("correlation of age with other columns\n",df.corr(numeric_only=True)['Age'])
```
